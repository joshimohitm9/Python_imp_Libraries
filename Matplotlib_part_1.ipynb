**Basics of Matplotlib**

```python
# importing library
import  matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.arange(1,10)
a
```

```python
y1=a*2
y2=a*3
```

```python
y3=a*3
y4=a*4
```

```python
# creating a plot
plt.plot(y1,y2,color="red",marker="o",linestyle="dotted",linewidth=2)
```

```python
# creating  a subplot

plt.subplot(1,2,1)
plt.plot(y1,y2,color="red",marker="o",linestyle="dotted",linewidth=2)
plt.subplot(1,2,2)
plt.plot(y3,y4,color="pink",marker="*",linestyle="dotted",linewidth=5)
plt.xlabel("mohit")
plt.ylabel("joshi")
plt.title("MOHIT")
plt.grid()
plt.show()
```

```python
# barplot
plt.bar(y3,y4,width=2)
# for grid line
plt.grid()
plt.show()
```

```python
# horizontal bar chart
plt.barh(y3,y4,height=2)
plt.grid()
plt.show()
```

```python
# scatter plot
plt.scatter(y1,y3,marker="o",s=100)
plt.grid()
plt.plot()
```

```python
# box plot
plt.boxplot([y1,y3])
plt.show()
```

```python
# volin plot
plt.violinplot([y1,y3])
plt.show()
```

```python
fruits=["mango","twwinkle","banana","apple"]
kg=[10,50,20,2]
```

```python
# creating a pie chart
plt.pie(kg,labels=fruits,colors=["orange","pink","yellow","red"])
```

```python
# creating a donut chart
plt.pie(kg,labels=fruits,colors=["orange","pink","yellow","red"],radius=2)
plt.pie([1],colors=["w"],radius=1)
plt.show()
```
