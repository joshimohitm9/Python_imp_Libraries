```python
# creating a data frame
data={
    "population":[1092,1269,4631,4328],
    "hospital":[189,208,149,286],
    "school":[6759,9555,8757,6567]
}

df=pd.DataFrame(data)
df
```

```python
# Assigning the index  (row )value
df.index=["Delhi","Mumbai","Kolkata","Chennai"]
df
```

```python
# creating an another Column named as College
df["college"]=np.linspace(2500,4000,4)
df
```

```python
# for looking into 1 column
df.population
# or
# df["population"]
```

```python
# for more then 1 columns
df[["hospital","college"]]
```

```python
# sorting
df.sort_values(by="population",ascending=False)
# higher the population will shown above
```

```python
# set the value of first row as 1000 using loc function
df.loc["Kolkata",:]=1000
df
```

```python
# this is used to drop the columns name college from axis =1
df.drop("college",axis=1,inplace=True)
```

```python
# help to reset the index
df.reset_index()
```

```python
# filter where population is greater then 2000
df[df["population"]>2000]
```

```python
df[df["hospital"]>200]
```

```python
# created a column name cat where if the school is gretaer then 1000 then yes else no
df["cat"]=["yes" if i>1000 else "No" for i in df["school"]]
```
