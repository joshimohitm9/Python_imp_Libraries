```python
# using seaborn library we are creating a line plot
A=sns.lineplot(x="total_rooms",y="total_bedrooms",data=a,marker="o")
```

```python
# scatteer plot
sns.scatterplot(x="longitude",y="latitude",data=a)
```

```python
# changing the datatype of ""housing_median_age"" column
a["housing_median_age"]=a["housing_median_age"].astype(int)


# if ""housing_median_age"" is greater the 30 then True else false
a["newcol"]= a["housing_median_age"]>30
```

```python
a.head()
```

```python
# hue parameter
sns.scatterplot(x="longitude",y="latitude",data=a,hue="newcol")
```

```python
x=a["longitude"].head(20)
y=a["latitude"].head(20)
```

```python
sns.lineplot(x=x,y=y,data=a,hue="newcol")
```

```python
# barplot
sns.barplot(x=x,y=y,data=a,hue="newcol")
```

```python
# barplot
import matplotlib.pyplot as plt

# for changing the size of figure
plt.figure(figsize=(15,5))

sns.barplot(x=x,y=y,data=a,hue="newcol")
```

```python
# for labeling on the top of bar
plt.figure(figsize=(15,5))

m=sns.barplot(x=x,y=y,data=a,hue="newcol")

for bars in m.containers:
  m.bar_label(bars)
```

```python
sns.histplot(x="longitude",y="latitude",data=a,hue="newcol")
```
